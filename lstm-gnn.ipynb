### Use the average of that hour of that day of the week (of the training set) to fill the nulls of both sets

```python

sensor_columns = [col for col in df.columns if col not in ['hour', 'day_of_week']]

train = train.copy()


# Iterate through each row
def fill_df(df_to_fill):
    for idx, row in df_to_fill.iterrows():
        hour = idx.hour
        day_of_week = idx.dayofweek

        # Iterate through each sensor column
        for col in sensor_columns:
            # Check if the value is missing
            if pd.isna(row[col]):
                # Replace with the average for that day of week and hour
                avg_value = hourly_avg.loc[(day_of_week, hour), col]
                df_to_fill.at[idx, col] = avg_value


fill_df(train)
fill_df(test)

print(train.isnull().sum())
print(test.isnull().sum())
df[["773869"]].plot(figsize=(12, 6))
train[["773869"]].plot(figsize=(12, 6))
```

```python
import numpy as np
import pandas as pd

# Assume train_tf is already loaded as a DataFrame with timestamps as the index.
# For example:
# train_tf = pd.read_csv('your_file.csv', index_col=0)

# Specify the auxiliary columns that should not be used in the computation.
auxiliary_cols = ['hour', 'day_of_week']

# Filter out auxiliary columns to obtain only sensor data.
sensor_columns = [col for col in train_tf.columns if col not in auxiliary_cols]
sensor_data = train_tf[sensor_columns]

num_sensors = sensor_data.shape[1]

# Compute the pairwise distance matrix based on Pearson correlation.
# We define the distance as: distance = 1 - correlation.
distance_matrix = np.zeros((num_sensors, num_sensors))
for i in range(num_sensors):
    for j in range(num_sensors):
        if i != j:
            corr = np.corrcoef(sensor_data.iloc[:, i], sensor_data.iloc[:, j])[0, 1]
            distance_matrix[i, j] = 1 - corr
        else:
            distance_matrix[i, j] = 0

# Function to compute the adjacency matrix using a Gaussian kernel and threshold.
def compute_adjacency_matrix(distance_matrix: np.ndarray, sigma2: float, epsilon: float):
    num_sensors = distance_matrix.shape[0]
    w2 = distance_matrix * distance_matrix
    w_mask = np.ones((num_sensors, num_sensors)) - np.eye(num_sensors)
    adjacency_matrix = (np.exp(-w2 / sigma2) >= epsilon) * w_mask
    return adjacency_matrix

sigma2 = 0.1
epsilon = 0.5

# Compute the adjacency matrix.
adjacency_matrix = compute_adjacency_matrix(distance_matrix, sigma2, epsilon)

# Extract edges from the adjacency matrix.
node_indices, neighbor_indices = np.where(adjacency_matrix == 1)

# Define the GraphInfo class.
class GraphInfo:
    def __init__(self, edges: tuple, num_nodes: int):
        self.edges = edges
        self.num_nodes = num_nodes

# Build the graph from the computed edges.
graph = GraphInfo(
    edges=(node_indices.tolist(), neighbor_indices.tolist()),
    num_nodes=adjacency_matrix.shape[0],
)

print(f"Number of sensors: {graph.num_nodes}, number of edges: {len(graph.edges[0])}")
```

```python
import typing 
import tensorflow as tf
import keras
from keras import layers
from keras import ops
```

```python
class GraphConv(layers.Layer):
    def __init__(
        self,
        in_feat,
        out_feat,
        graph_info: GraphInfo,
        aggregation_type="mean",
        combination_type="concat",
        activation: typing.Optional[str] = None,
        **kwargs,
    ):
        super().__init__(**kwargs)
        self.in_feat = in_feat
        self.out_feat = out_feat
        self.graph_info = graph_info
        self.aggregation_type = aggregation_type
        self.combination_type = combination_type
        self.weight = self.add_weight(
            initializer=keras.initializers.GlorotUniform(),
            shape=(in_feat, out_feat),
            dtype="float32",
            trainable=True,
        )
        self.activation = layers.Activation(activation)

    def aggregate(self, neighbour_representations):
        aggregation_func = {
            "sum": tf.math.unsorted_segment_sum,
            "mean": tf.math.unsorted_segment_mean,
            "max": tf.math.unsorted_segment_max,
        }.get(self.aggregation_type)

        if aggregation_func:
            return aggregation_func(
                neighbour_representations,
                self.graph_info.edges[0],
                num_segments=self.graph_info.num_nodes,
            )

        raise ValueError(f"Invalid aggregation type: {self.aggregation_type}")

    def compute_nodes_representation(self, features):
        """Computes each node's representation.

        The nodes' representations are obtained by multiplying the features tensor with
        `self.weight`. Note that
        `self.weight` has shape `(in_feat, out_feat)`.

        Args:
            features: Tensor of shape `(num_nodes, batch_size, input_seq_len, in_feat)`

        Returns:
            A tensor of shape `(num_nodes, batch_size, input_seq_len, out_feat)`
        """
        return ops.matmul(features, self.weight)

    def compute_aggregated_messages(self, features):
        neighbour_representations = tf.gather(features, self.graph_info.edges[1])
        aggregated_messages = self.aggregate(neighbour_representations)
        return ops.matmul(aggregated_messages, self.weight)

    def update(self, nodes_representation, aggregated_messages):
        if self.combination_type == "concat":
            h = ops.concatenate([nodes_representation, aggregated_messages], axis=-1)
        elif self.combination_type == "add":
            h = nodes_representation + aggregated_messages
        else:
            raise ValueError(f"Invalid combination type: {self.combination_type}.")
        return self.activation(h)

    def call(self, features):
        """Forward pass.

        Args:
            features: tensor of shape `(num_nodes, batch_size, input_seq_len, in_feat)`

        Returns:
            A tensor of shape `(num_nodes, batch_size, input_seq_len, out_feat)`
        """
        nodes_representation = self.compute_nodes_representation(features)
        aggregated_messages = self.compute_aggregated_messages(features)
        return self.update(nodes_representation, aggregated_messages)
```

```python
class LSTMGC(layers.Layer):
    """Layer comprising a convolution layer followed by LSTM and dense layers."""

    def __init__(
        self,
        in_feat,
        out_feat,
        lstm_units: int,
        input_seq_len: int,
        output_seq_len: int,
        graph_info: GraphInfo,
        graph_conv_params: typing.Optional[dict] = None,
        **kwargs,
    ):
        super().__init__(**kwargs)

        # graph conv layer
        if graph_conv_params is None:
            graph_conv_params = {
                "aggregation_type": "mean",
                "combination_type": "concat",
                "activation": None,
            }
        self.graph_conv = GraphConv(in_feat, out_feat, graph_info, **graph_conv_params)
        self.lstm = layers.LSTM(lstm_units, activation="relu")
        self.dense = layers.Dense(output_seq_len)

        self.input_seq_len, self.output_seq_len = input_seq_len, output_seq_len

    def call(self, inputs):
        """Forward pass.

        Args:
            inputs: tensor of shape `(batch_size, input_seq_len, num_nodes, in_feat)`

        Returns:
            A tensor of shape `(batch_size, output_seq_len, num_nodes)`.
        """

        # convert shape to  (num_nodes, batch_size, input_seq_len, in_feat)
        inputs = ops.transpose(inputs, [2, 0, 1, 3])

        gcn_out = self.graph_conv(
            inputs
        )  # gcn_out has shape: (num_nodes, batch_size, input_seq_len, out_feat)
        shape = ops.shape(gcn_out)
        num_nodes, batch_size, input_seq_len, out_feat = (
            shape[0],
            shape[1],
            shape[2],
            shape[3],
        )

        # LSTM takes only 3D tensors as input
        gcn_out = ops.reshape(
            gcn_out, (batch_size * num_nodes, input_seq_len, out_feat)
        )
        lstm_out = self.lstm(
            gcn_out
        )  # lstm_out has shape: (batch_size * num_nodes, lstm_units)

        dense_output = self.dense(
            lstm_out
        )  # dense_output has shape: (batch_size * num_nodes, output_seq_len)
        output = ops.reshape(dense_output, (num_nodes, batch_size, self.output_seq_len))
        return ops.transpose(
            output, [1, 2, 0]
        )  # returns Tensor of shape (batch_size, output_seq_len, num_nodes)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
def create_dataset(df, input_sequence_length=12, forecast_horizon=3, batch_size=64):
    """
    Creates a TensorFlow dataset for the LSTMGC model.
    
    Args:
        df: DataFrame with shape (time_steps, num_nodes)
        input_sequence_length: Number of time steps in each input sequence
        forecast_horizon: Number of time steps to predict
        batch_size: Batch size for training
    
    Returns:
        tf.Dataset with features of shape (batch_size, input_sequence_length, num_nodes, 1)
                      and targets of shape (batch_size, forecast_horizon, num_nodes)
    """
    # Convert dataframe to numpy array
    data = df.values  # Shape: (time_steps, num_nodes)
    
    # Create sequences
    sequences = []
    targets = []
    
    for i in range(len(data) - input_sequence_length - forecast_horizon + 1):
        # Input sequence
        seq = data[i:i+input_sequence_length]  # Shape: (input_sequence_length, num_nodes)
        # Target sequence (next forecast_horizon time steps)
        target = data[i+input_sequence_length:i+input_sequence_length+forecast_horizon]  # Shape: (forecast_horizon, num_nodes)
        
        sequences.append(seq)
        targets.append(target)
    
    # Convert to numpy arrays
    X = np.array(sequences)  # Shape: (num_sequences, input_sequence_length, num_nodes)
    y = np.array(targets)    # Shape: (num_sequences, forecast_horizon, num_nodes)
    
    # Reshape X to add the feature dimension
    X = X.reshape(X.shape[0], X.shape[1], X.shape[2], 1)  # Shape: (num_sequences, input_sequence_length, num_nodes, 1)
    
    # Create TensorFlow dataset
    dataset = tf.data.Dataset.from_tensor_slices((X, y))
    dataset = dataset.shuffle(buffer_size=len(X))  # Shuffle with a buffer size
    dataset = dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    
    return dataset
```

```python
def create_validation_test_split(test_df, validation_ratio=0.5):
    """
    Split the test dataframe into validation and test sets while preserving temporal order.
    
    Args:
        test_df: DataFrame containing the test data
        validation_ratio: Portion of test_df to use for validation (0 to 1)
    
    Returns:
        validation_df, test_df: The split dataframes
    """
    # Calculate the split point
    split_idx = int(len(test_df) * validation_ratio)
    
    # Split the data
    validation_df = test_df.iloc[:split_idx]
    final_test_df = test_df.iloc[split_idx:]
    
    return validation_df, final_test_df

# Split the test data
validation_tf, test_tf_final = create_validation_test_split(test_tf)

print(f"Original test set size: {len(test_tf)}")
print(f"Validation set size: {len(validation_tf)}")
print(f"Final test set size: {len(test_tf_final)}")
```

```python
from sklearn.metrics import mean_squared_error

def calculate_true_mse(model, dataset, scaler):
    """
    Calculate MSE on the original scale by reversing the StandardScaler transformation.
    
    Args:
        model: Trained LSTMGC model
        dataset: tf.Dataset with scaled input sequences
        scaler: The StandardScaler used to scale the original data
    
    Returns:
        The MSE on the original scale
    """
    true_targets = []
    true_predictions = []
    
    # Get predictions and targets from the dataset
    for X_batch, y_batch in dataset:
        # Make predictions
        y_pred = model.predict(X_batch, verbose=0)
        
        # For each sample in the batch
        for i in range(len(y_batch)):
            # Get scaled target and prediction
            scaled_target = y_batch[i].numpy()  # Shape: (forecast_horizon, num_nodes)
            scaled_pred = y_pred[i]  # Shape: (forecast_horizon, num_nodes)
            
            # Reshape for inverse transform
            # The scaler expects shape (n_samples, n_features)
            scaled_target_reshaped = scaled_target.reshape(-1, scaled_target.shape[-1])
            scaled_pred_reshaped = scaled_pred.reshape(-1, scaled_pred.shape[-1])
            
            # Inverse transform
            original_target = scaler.inverse_transform(scaled_target_reshaped)
            original_pred = scaler.inverse_transform(scaled_pred_reshaped)
            
            # Store the values
            true_targets.append(original_target)
            true_predictions.append(original_pred)
    
    # Concatenate all predictions and targets
    true_targets = np.concatenate(true_targets)
    true_predictions = np.concatenate(true_predictions)
    
    # Calculate MSE on the original scale
    true_mse = mean_squared_error(true_targets, true_predictions)
    
    return true_mse
```

```python
in_feat = 1
n_layers = 4
batch_size = 128
epochs = 150
input_sequence_length = 16
forecast_horizon = 6
out_feat = 32
lstm_units = 64
graph_conv_params = {
    "aggregation_type": "mean",
    "combination_type": "concat",
    "activation": None,
}

model = Sequential()
model.add(layers.Input((input_sequence_length, graph.num_nodes, in_feat)))
for k in range(n_layers):
    # First layer takes input shape (batch_sz, input_sequence_length, graph.num_nodes, in_feat)
    if k == 0:
        model.add(LSTMGC(
            in_feat,
            out_feat,
            lstm_units,
            input_sequence_length,
            forecast_horizon,
            graph,
            graph_conv_params,
        ))  # returns Tensor of shape (batch_size, output_seq_len, num_nodes)
    else:
        # Reshape/expand dimensions before feeding to next layer
        model.add(tf.keras.layers.Reshape((forecast_horizon, graph.num_nodes, 1)))
        # Subsequent layers take the expanded output as input
        model.add(LSTMGC(
            1,  # expanded dimension
            out_feat,
            lstm_units,
            forecast_horizon,  # previous output sequence length becomes input length
            forecast_horizon,
            graph,
            graph_conv_params,
        ))
model.summary()
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss=keras.losses.MeanSquaredError(),
)


# Define a directory to save the best model
checkpoint_dir = './model_checkpoints'
os.makedirs(checkpoint_dir, exist_ok=True)
model_checkpoint = keras.callbacks.ModelCheckpoint(
        filepath=os.path.join(checkpoint_dir, 'best_model.weights.h5'),
        monitor='val_loss',
        save_best_only=True,
        save_weights_only=True,
        verbose=1
    )
history = model.fit(
    create_dataset(train_tf[sensor_columns], input_sequence_length=input_sequence_length, forecast_horizon=forecast_horizon, batch_size=batch_size),
    validation_data=create_dataset(validation_tf[sensor_columns], input_sequence_length=input_sequence_length, forecast_horizon=forecast_horizon, batch_size=batch_size),
    epochs=epochs,
    callbacks=[keras.callbacks.EarlyStopping(patience=20), model_checkpoint],
)

# Extract the MSE values
train_loss = history.history['loss']  # This will be MSE since you're using MSE loss
val_loss = history.history['val_loss']

# Create figure
plt.figure(figsize=(10, 6))

# Plot training & validation MSE
plt.plot(train_loss, label='Training MSE', linewidth=2)
plt.plot(val_loss, label='Validation MSE', linewidth=2)

# Add labels and title
plt.title('MSE over epochs', fontsize=14)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Mean Squared Error', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)

# Add epoch markers
epochs = range(1, len(train_loss) + 1)
best_epoch = val_loss.index(min(val_loss)) + 1
plt.axvline(x=best_epoch, color='r', linestyle='--', alpha=0.5)
plt.text(best_epoch + 0.5, min(val_loss), f'Best epoch: {best_epoch}', 
         verticalalignment='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
